```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

# Business case for an offer from a startup

* See corresponding section in Chapter 5

```python
def discount_rate(i,T):
    '''
    Compute discount rate given base anual inflation (i) and maturity (T)
    '''
    r_it = (1+i)**T - 1
    return r_it
```

```python
def npv_offer(wage_offer,discount_rate,stocks,price_stocks):
    '''
    Compute the simplified (2-steps) npv of the offer
    '''
    value_equity = stocks*price_stocks
    npv = wage_offer + (wage_offer + value_equity)/(1+discount_rate)
    return npv
```

```python
def npv_base(wage_offer,discount_rate):
    '''
    Compute the simplified (2-steps) npv of base salary
    '''
    npv = wage_offer + wage_offer/(1+discount_rate)
    return npv
```

```python
# Check discount_rate function
inflation_rate = 0.04
periods_vesting = 4
print(discount_rate(i=inflation_rate, T=periods_vesting))
```

```python
# Check npv functions
wage_base = 100
wage_reduction = 0.8
wage_offer= wage_base*wage_reduction
my_discount_rate = discount_rate(i=inflation_rate, T=periods_vesting)
stocks = 100
price_stocks = 0.2
offer_value = npv_offer(wage_offer, my_discount_rate, stocks, price_stocks)
base_value = npv_base(wage_base, my_discount_rate)
print('Value of the job offer = {0}'.format(offer_value))
print('Value of current job = {0}'.format(base_value))
```

### Compute maximum wage reduction you're willing to accept

```python
# Plot for different wages
wage_red_grid = 100*(1-np.linspace(0.8,1,40))
reduction_df = pd.DataFrame(index=wage_red_grid, columns=['pct'])
for w,red in enumerate(wage_red_grid):
    offer_red = wage_base*(1-red/100)
    offer_value_r = npv_offer(offer_red, my_discount_rate, stocks, price_stocks)
    reduction_df.pct.loc[red] = offer_value_r
    
fig, ax = plt.subplots(figsize=(10,6))
ax.plot(reduction_df.index, reduction_df.pct, color='k')
ax.plot([0,20],[base_value,base_value], color='r')
ymin,ymax=ax.get_ylim()
ax.axis([0,20,ymin,ymax])
ax.set_title('Finding the Maximum Wage Reduction', fontsize=18)
ax.set_ylabel('$', fontsize=14)
ax.set_xlabel('% Wage Reduction', fontsize=16)
base_value_str = '$'+str(np.around(base_value,decimals=0))[:3]
ax.text(2.5,base_value,'Current Base = {0}'.format(base_value_str), verticalalignment='top', color='r', fontsize=14)
ax.text(2.0,200,'Offers for given wage reduction', fontsize=14)
# plot max wage reduction: find the intersection of both lines
max_wage_reduction = np.abs(reduction_df.pct-base_value).astype('float').idxmin()
ax.plot([max_wage_reduction, max_wage_reduction],[ymin, base_value], ls='--', color='k')
max_wage_str = '{0}%'.format(np.around(max_wage_reduction,decimals=1))
ax.text(max_wage_reduction,170,max_wage_str, horizontalalignment='right',
       fontsize=12)
print('')
```
